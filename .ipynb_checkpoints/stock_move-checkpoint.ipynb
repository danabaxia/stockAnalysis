```python
#import library
import financial as f
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#load data
tker = 'AAPL'
X = f.requestHistoryStockPriceByDay(tker,300)
Y = []
Date = []
for item in X:
    Y.append(item['adjClose'])
    Date.append(item['date'])
Data = pd.DataFrame(Y[::-1])
Data
```

```python
#Visualize the data

plt.figure(figsize=(12.5,4.5))
plt.plot(Data,label = 'AAPL')
plt.title('history price')
plt.xlabel('history')
plt.ylabel('price')
plt.legend(loc='upper left')
plt.show()

```

```python
#create ave move 
SMA50 = pd.DataFrame()
SMA50 = Data.rolling(window=50).mean()
SMA10 = pd.DataFrame()
SMA10 = Data.rolling(window=10).mean()
SMA30 = pd.DataFrame()
SMA30 = Data.rolling(window=30).mean()
SMA5 = pd.DataFrame()
SMA5 = Data.rolling(window=5).mean()
```

```python
#visualize the data
plt.figure(figsize=(12.5,4.5))
plt.plot(Data,label = 'price')
plt.plot(SMA50,label = 'SMA50')
plt.plot(SMA30,label = 'SMA30')
plt.plot(SMA10,label = 'SMA10')
plt.plot(SMA5,label = 'SMA5')
plt.title(tker)
plt.xlabel('history')
plt.ylabel('price')
plt.legend(loc='upper left')
plt.show()
```
